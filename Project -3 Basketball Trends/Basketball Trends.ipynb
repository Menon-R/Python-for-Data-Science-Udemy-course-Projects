```python
#Creating a function to loop through the players and make visualizations
def myplot(data, playerlist = Players):
    
    col={"LeBronJames":"black","PaulGeorge":"red","KevinDurant":"blue","ChrisPaul":"green","RussellWestbrook":"yellow"}
    mrk={"LeBronJames":"o","PaulGeorge":"s","KevinDurant":"^","ChrisPaul":"D","RussellWestbrook":"s"}
    
    for name in playerlist:
        plt.plot(data[Pdict[name]],c=col[name],ls='--',marker=mrk[name],ms='7',label=name)
    plt.legend(loc='upper left',bbox_to_anchor=(1,1))
    plt.xticks(list(range(0,6)),Seasons,rotation='vertical')
    plt.show()
```

```python
#Plot for Player Salary Yr 2014-19
myplot(Salary)
```
